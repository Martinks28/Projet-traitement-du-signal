(2) $$S(x,t) = f(x + \Delta(t))$$ où $$\Delta(t) = \delta(t) + \alpha h \star \delta(t)$$ $$\mathcal{F}(x \mapsto S(x,t))(\omega) = e^{i\Delta(t)\omega}\hat{f}(\omega)$$ car une translation dans l'espace physique = décalage de phase en fréquentiel  

(3) $$S(x,t) = f(x) + \Delta(t)\frac{\partial f}{\partial x}(x)$$ $$\mathcal{F}(x \mapsto S(x,t))(\omega) = \hat{f}(\omega) + \Delta(t)\widehat{\frac{\partial f}{\partial x}}(\omega)$$ $$\widehat{\frac{\partial f}{\partial x}}(\omega) = \frac{1}{2\pi} \int_{\mathbb{R}} \frac{\partial f}{\partial x} e^{-i\omega x} dx$$ $$= \frac{1}{2\pi} \underbrace{\left[ f(x) e^{-i\omega x} \right]_{\mathbb{R}}}_{0 \text{ car f est à support compact}} + \frac{1}{2\pi} i\omega \int_{\mathbb{R}} f(x) e^{-i\omega x} dx$$ $$= i\omega \hat{f}(\omega)$$ donc $$\mathcal{F}(x \mapsto S(x,t))(\omega) = \hat{f}(\omega)(1 + i\omega \Delta(t))$$
